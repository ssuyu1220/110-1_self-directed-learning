```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-360,360,5000)
y1=np.sin(x*np.pi/180)
y2=np.cos(1/3*x*np.pi/180-(np.pi/180)*180)
y=y1-y2
plt.xlabel('x axis (degree)')
plt.ylabel('y axis')
plt.grid(True)
plt.plot(x,y,'--',color='purple',linewidth=2.5,label='sinx-cos((1/3)*x-180$^\circ$)')
leg = plt.legend(loc='lower left') 

```
